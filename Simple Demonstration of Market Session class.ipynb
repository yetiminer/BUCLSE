This is how you run one session of 600 time units. Information will be spat out into a csv named specified by 'trade_file'

```python
#define the market_session object
environ_dic['verbose']=True
sess=Market_session(**environ_dic)
#run the simulation
sess.simulate()
```

The original method repeatedly opened and added to a csv file. I thought this was a bit crude so wrote an update to use dataframes instead and write the file only at the end. 

```python
sess._reset_session()
sess.trade_file='trade_file_pandas.csv'
sess.simulate(sess.trade_stats_df3)
```
